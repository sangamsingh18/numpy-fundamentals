```python
import numpy as np

# 1D Arrays
a = np.array([1, 2])
b = np.array([3, 4])

# Vertical Stack (Top to Bottom)
v_result = np.vstack((a, b))
print("Vertical Stack (vstack):")
print(v_result)

# Horizontal Stack (Side by Side)
h_result = np.hstack((a, b))
print("\nHorizontal Stack (hstack):")
print(h_result)
```
